# Capstone - Battle of Neighbourhoods. What is the best hotel/accommodation when visiting Singapore?

## Business problem

Through this project, I will identify the optimal rental location in Singapore for someone looking for an apartment, based on price and proximity to places of interest (in particular food venues, which are very important). I will analyse specific neighbourhoods using the Foursquare API. 

Although it is a small nation-state, Singapore is a very expensive place to live in, with small apartments going for millions of dollars. Apartments for rent are also expensive, even in the short term. This project is thus important to help renters find the best value for money rental.

The target audience is short-term renters, e.g. people who have moved here for work. The goal is to recommend an optimal housing location to rent.

## Data required

I will need: 
* Singapore towns and median residential rental prices (available from Singapore government open dataset 
* Most common venues in neighbourhoods in Singapore
* Characteristics of housing available to purchase, e.g. price

Data sources: 
* Most common venues in neighbourhoods - from Foursquare API 
* Pandsas ython library for neighbourhood data from Wikipedia
