# Capstone - Battle of Neighbourhoods. Where to rent in Singapore?

## Intro/Business Problem

Through this project, I will identify the optimal rental location in Singapore for someone looking for an apartment, based on price and proximity to places of interest (in particular food venues, which are very important). I will analyse specific neighbourhoods using the Foursquare API. 

Although it is a small nation-state, Singapore is a very expensive place to live in, with small apartments going for millions of dollars. Apartments for rent are also expensive, even in the short term. This project is thus important to help renters find the best value for money rental.

The target audience is short-term renters, e.g. people who have moved here for work. The goal is to recommend an optimal housing location to rent.

## Data Required and Data Sources

I will be using the following data sources: 
* Singapore towns and median residential rental prices (available from Singapore government open dataset https://data.gov.sg/dataset/b35046dc-7428-4cff-968d-ef4c3e9e6c99) 
* Singapore town locations using Google Maps API
* Singapore food venue recommendations using Foursquare API (limited to 50 queries per day)

These will help us to analyse important factors like price and proximity to good food, to determine the best rental location.

## Methodology 

First, find the data required (see above) for rental prices.
Second, clean the data. Drop missing data, eliminate duplicates, fix data types. Also use the most current record as of 2020. 
Third, get the coordinates using Google APIs. This is to help find FourSquare API locations.
Fourth, get nearby venues from FourSquare. In this example, food venues were used. 
* kmeans clustering was used to group neighbourhoods into clusters, based on their most common venue categories.

Please see the github entry for the code used. 

## Results

The results showed that the best locations for price were further out - Bukit Panjang, Woodlands, Sembawang (north and west) all had the lowest rent. This makes sense as their locations are not as good. However, the Marine Parade area has low rent and also good food around it, especially many food courts. 


## Discussion

The results showed that it was possible to find a decently priced area to rent that was not too far away from the downtown district, but still have good food. This is slightly surprising as most people think that one can either get good food and good location, or good price. 

## Conclusion

People looking to rent a place in Singapore and who are conscious of price and good food should consider the Marine Parade neighbourhood. It has won the 'battle of the neighbourhoods'.
